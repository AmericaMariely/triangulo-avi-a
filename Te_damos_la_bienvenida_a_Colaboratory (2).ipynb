```
class AnalizadorCalificaciones:
    def __init__(self):
        self.nombres = []
        self.calificaciones = []
        self.promedio_grupal = 0.0
        self.alumnos_destacados = []

    def leer_datos(self):
        """Método para leer los nombres y calificaciones de 10 alumnos"""
        print("=== INGRESO DE DATOS DE ALUMNOS ===\n")

        for i in range(10):
            print(f"Alumno {i+1}:")
            nombre = input("Ingrese el nombre del alumno: ")
            calificacion = float(input("Ingrese la calificación del alumno: "))

            self.nombres.append(nombre)
            self.calificaciones.append(calificacion)
            print()

    def realizar_proceso(self):
        """Método para calcular el promedio grupal y alumnos destacados"""
        # Calcular promedio grupal
        self.promedio_grupal = sum(self.calificaciones) / len(self.calificaciones)

        # Obtener alumnos con calificación mayor a 9.5
        self.alumnos_destacados = []
        for i in range(len(self.nombres)):
            if self.calificaciones[i] > 9.5:
                self.alumnos_destacados.append({
                    'nombre': self.nombres[i],
                    'calificacion': self.calificaciones[i]
                })

    def imprimir_resultados(self):
        """Método para imprimir los resultados del análisis"""
        print("\n" + "="*50)
        print("RESULTADOS DEL ANÁLISIS ACADÉMICO")
        print("="*50)

        # Mostrar lista completa de alumnos
        print("\n--- LISTA COMPLETA DE ALUMNOS ---")
        for i in range(len(self.nombres)):
            print(f"{i+1:2d}. {self.nombres[i]:20} - Calificación: {self.calificaciones[i]:.2f}")

        # Mostrar promedio grupal
        print(f"\n--- PROMEDIO GRUPAL ---")
        print(f"Promedio del grupo: {self.promedio_grupal:.2f}")

        # Mostrar alumnos destacados
        print(f"\n--- ALUMNOS DESTACADOS (Calificación > 9.5) ---")
        if self.alumnos_destacados:
            for i, alumno in enumerate(self.alumnos_destacados, 1):
                print(f"{i}. {alumno['nombre']:20} - Calificación: {alumno['calificacion']:.2f}")
        else:
            print("No hay alumnos con calificación mayor a 9.5")

        print("="*50)

    def ejecutar_analisis(self):
        """Método principal que coordina todo el proceso"""
        self.leer_datos()
        self.realizar_proceso()
        self.imprimir_resultados()

# Ejecutar el programa
if __name__ == "__main__":
    analizador = AnalizadorCalificaciones()
    analizador.ejecutar_analisis()
```
